### Explain Columns:

* page_total_likes: Total number of people who liked/followed the company's page at the time the post was published (measures the reach potential).
* post_type: The format of the content (Photo, Status, Link, Video).
* category: The intent/theme of the post:
1. (Action): Focused on conversion (special offers, contests, coupons).
2. (Product): Direct advertisement or product showcase.
3. (Inspiration): Non-explicit, brand-related content meant to build community.
* post_month: The month of publication (1-12).
* post_weekday: The day of the week (1-7).
* post_hour: The hour of the day (1-24) the post was published.
* paid: Binary indicator (1: Yes, 0: No) of whether the company paid Facebook to promote/boost the post.


2. Reach & Impression Metrics (Exposure)
* lifetime_post_reach: The number of unique users who saw the post in their newsfeed or on the page.
* lifetime_post_impressions: The total number of times the post was displayed. If one person saw it three times, it counts as 1 for Reach and 3 for Impressions.
* lifetime_post_impressions_from_page_likes: Total views specifically generated by people who already follow the page.
* lifetime_post_reach_from_page_likes: Total number of unique followers who saw the post.

3. Engagement Metrics (User Behavior)
* lifetime_engaged_users: Unique people who clicked anywhere on the post (includes likes, comments, shares, but also "passive" clicks like viewing the profile or clicking "see more").
* lifetime_post_consumers: Unique people who clicked to consume the content specifically (viewing a photo, clicking a link, or playing a video).
* lifetime_post_consumptions: Total number of clicks on the content. One "Consumer" can generate multiple "Consumptions" by clicking several times.
* page_likes_engaged_users: The number of unique people who both follow the page and interacted with the post.

4. Interaction Metrics (The Target)
* post_comments: Total comments received.
* post_likes: Total "Like" reactions received.
* post_shares: Total times the post was shared.
* total_interactions (Target): The sum of likes, comments, and shares. This represents the "Public Engagement" score.

#### Concolusion of HeatMap
1. We have to remove `multicollinearity` correlations and there is many in our Case
2. We have Used Both Spearman for (Non-Linearity) and Pearson for (Linear Correlation)

##### Pearson & SpearMan 
1. Lifetime_post_reach / Lifetime_post_impression
2. Lifetime_post_impression_from_page_likes / lifetime_post_reach_from_page_likes
3. lifetime_engaged_users / lifetime_post_consumers

* Those are Must be Dropped or we have to create some `Feature Engineering` All of them Correlation is between `0.95-0.99`

```python
df['post_ratio_imp_reach'] = df['lifetime_post_impressions'] / df['lifetime_post_reach']
df['post_ratio_imp_rach_likes'] = df['lifetime_post_impressions_from_page_likes'] / df['lifetime_post_reach_from_page_likes']
df['post_ratio_engaged_consumers'] = df['lifetime_engaged_users'] / df['lifetime_post_consumers']
df['post_ratio_engaged_consumption'] = df['lifetime_post_consumptions'] / df['page_likes_engaged_users']
```
