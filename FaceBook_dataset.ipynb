### Explain Columns:

* page_total_likes: Number of people who have liked (Followed) the company's page
* post_type: Post can be Categorized into (`Photo`, `Status`, `Link`, `Video`)
* category: it put each post in a certain Category (1:action `special offers and contests`, 2: product`direct aderisment` 3: inspiration `non-explicit brand related content` 
* post_month: at which month did the user upload the post (1-12)
* post_weekday: at which day of the week did the user upload post (1-7)
* post_hour: which hour did user upload post (it ranges from 1-24, not 1-12)
* paid: did the company paid to promote the post (Yes, No)
* lifetime_post_reach: Number of people who saw a page post(unique users), impressions are the number of times a page from a page is displayed, whether the post is clicked or not, people may see multiple
* lifetime_post_impressions: impression of the same post for example someone might see a Page update in posts once and then a friend share it agian (non-unique) same as post_reach but non-unique
* lifetime_engaged_users: Number of people who clicked anywhere in a post (click, comment, likes)(unique users)
* lifetime_post_consumers:The number of people who clicked anywhere in post (just click or opend, doesn't require to like or comment) (unique)
* lifetime_post_consumptions: The number of clicks anywhere in a post (same as consumer but consumption (not unique))
* lifetime_post_impressions_from_page_likes: Total Number of impressions just from peopel who have followed page
* lifetime_post_reach_from_page_likes: The number of peple who saw a page post because they have followed that page ( unique users )
* page_likes_engaged_users: Number of people who followed page and clicked anywhere in a post (Uniqeue Users)
* post_comments: total number of comments received by the post
* post_likes: total number of likes received by the post
* post_shares: total number of times the post was shared
* total_interactions(Target): Total Number of interactions on the post ( sum of (comments, likes, shares))

### Data Analyzing

#### Page Total likes

```python
"""
we can Categorize it into 
- High (136393 < x)
- Medium (136393 > x & x > 100000)
- Low (x < 100000)
"""
df.page_total_likes.describe()
df['liked_group'] = pd.qcut(df['page_total_likes'],q=3,labels=['Low','Medium','High'])
```

```python
"""
* This plot shows that High Page followers has a perfect distribution in their posts
* in other hand Medium ones mean isn't that good but the maximum is (it could be an outlier)
* Low: has a Perfect mean and maximum has some Questions we have to explore
"""
sns.violinplot(x='liked_group',y=df.total_interactions,data=df)
plt.yscale('log')
plt.show()
```

```python
"""
- Questions:
1- why there is some Pages with high followers and paid post, has a low likes?
2- 5000 could be an outlier
3- lets see much columns using groupby rather than plots
"""
sns.scatterplot(x='page_total_likes',y='post_likes',hue='paid',data=df)
```

```python
"""
Low:
[
1- As we see the huge difference between Paid and unpaid in term of Photo and status
{ paid: Photo with 42 accounts = 135 likes, Status: with 2 people = 328.5}
{unpaid: Photo 109 Accounts with 93 Likes, Status 3 Accounts = 55}
2- Paid one has A huge effect on the Post_likes
]
Medium:
[
1- A huge difference between paid and not paid esspicially on Video Status
{Paid: Status with 6 Accounts = 158, Video with 1 Account = 315}
{Unpaid: Status with 10 Accounts = 85, Video with 2 Accounts = 275}
2- so Differences in Those two Types is seenable
]
High:
[
1- The effection here is really High, it Higher Followers with Paid ones distroys teh Unpaid ones
2- Numbers here are talk about them selves
]
"""
df.groupby(['liked_group','paid','post_type'])['post_likes'].agg(['count','median'])
```

```python
size_liked_group = df.groupby(['liked_group'])['total_interactions'].mean().reset_index()
plt.title('Mean of Total interaction for each Liked group')
sns.barplot(x='liked_group',y='total_interactions',data=size_liked_group,hue='liked_group')
plt.show()
```

```python
"""
- Low has a certain months that Medium doesn't post in it and Medium has a certatin month where high doesn't touch it!!!
- there's a high correlation between them !!!
"""
sns.catplot(x= 'liked_group',y='post_month',data=df,hue='liked_group')
plt.title('Corr Between Post Month And Liked Group')
plt.show()
```
